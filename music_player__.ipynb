```python
import tkinter as tk
from tkinter import filedialog
import pygame
import os

# Initialize pygame mixer
pygame.mixer.init()

# Create the main window
root = tk.Tk()
root.title("Python Music Player")
root.geometry("600x400")
root.configure(bg="#2C3E50")  # Set background color

# Create a playlist (replace these with your actual song file paths)
playlist = [
    "C:/Users/Kavya/Music/Undiporaadhey - SenSongsMp3.Co.mp3",
    "C:/Users/Kavya/Music/[iSongs.info] 03 - Kavithe Kavithe.mp3",
    "C:/Users/Kavya/Music/[iSongs.info] 05 - Eppudo Ninnu.mp3",
]

# Create a favorites list
favorites = ["C:/Users/Kavya/Music/[iSongs.info] 01 - Katheyonda Helide.mp3"]

# Function to add a song to the playlist
def add_song():
    song = filedialog.askopenfilename(initialdir="your_music_directory/", filetypes=(("Audio Files", "*.mp3"),))
    if song:
        playlist.append(song)
        song_listbox.insert(tk.END, os.path.basename(song))

# Function to remove a song from the playlist
def remove_song():
    selected_song = song_listbox.curselection()
    if selected_song:
        selected_song = selected_song[0]
        del playlist[selected_song]
        song_listbox.delete(selected_song)

# Function to play a song
def play_song():
    selected_song = song_listbox.curselection()
    if selected_song:
        selected_song = selected_song[0]
        pygame.mixer.music.load(playlist[selected_song])
        pygame.mixer.music.play()

# Function to pause the currently playing song
def pause_song():
    pygame.mixer.music.pause()

# Function to add the currently playing song to favorites
def add_to_favorites():
    selected_song = song_listbox.curselection()
    if selected_song:
        selected_song = selected_song[0]
        song = playlist[selected_song]
        favorites.append(song)
        favorite_listbox.delete(0, tk.END)  # Clear the favorites listbox
        for fav_song in favorites:
            favorite_listbox.insert(tk.END, os.path.basename(fav_song))

# Function to play the next song
def play_next():
    selected_song = song_listbox.curselection()
    if selected_song:
        selected_song = selected_song[0]
        selected_song = (selected_song + 1) % len(playlist)
        pygame.mixer.music.load(playlist[selected_song])
        pygame.mixer.music.play()
        song_listbox.selection_clear(0, tk.END)
        song_listbox.activate(selected_song)
        song_listbox.selection_set(selected_song)

# Function to play the previous song
def play_previous():
    selected_song = song_listbox.curselection()
    if selected_song:
        selected_song = selected_song[0]
        selected_song = (selected_song - 1) % len(playlist)
        pygame.mixer.music.load(playlist[selected_song])
        pygame.mixer.music.play()
        song_listbox.selection_clear(0, tk.END)
        song_listbox.activate(selected_song)
        song_listbox.selection_set(selected_song)

# Function to set the volume
def set_volume(val):
    volume = int(val) / 100
    pygame.mixer.music.set_volume(volume)

# Create a frame for playlist controls
playlist_frame = tk.Frame(root, bg="#2C3E50")
playlist_frame.pack(pady=10)

add_button = tk.Button(playlist_frame, text="Add Song", command=add_song, bg="#3498DB", fg="#ECF0F1")
add_button.grid(row=0, column=0, padx=5)

remove_button = tk.Button(playlist_frame, text="Remove Song", command=remove_song, bg="#E74C3C", fg="#ECF0F1")
remove_button.grid(row=0, column=1, padx=5)

play_button = tk.Button(playlist_frame, text="Play", command=play_song, bg="#2ECC71", fg="#ECF0F1")
play_button.grid(row=0, column=2, padx=5)

pause_button = tk.Button(playlist_frame, text="Pause", command=pause_song, bg="#F39C12", fg="#ECF0F1")
pause_button.grid(row=0, column=3, padx=5)

next_button = tk.Button(playlist_frame, text="Next", command=play_next, bg="#3498DB", fg="#ECF0F1")
next_button.grid(row=0, column=4, padx=5)

previous_button = tk.Button(playlist_frame, text="Previous", command=play_previous, bg="#3498DB", fg="#ECF0F1")
previous_button.grid(row=0, column=5, padx=5)

# Create a frame for the playlist listbox
song_listbox_frame = tk.Frame(root, bg="#34495E")
song_listbox_frame.pack(pady=10)

song_listbox = tk.Listbox(song_listbox_frame, selectmode=tk.SINGLE, width=60, height=10, bg="#2C3E50", fg="#ECF0F1", selectbackground="#3498DB")
for song in playlist:
    song_listbox.insert(tk.END, os.path.basename(song))
song_listbox.pack()

# Create a frame for the favorites listbox
favorites_frame = tk.Frame(root, bg="#2C3E50")
favorites_frame.pack(pady=10)

add_favorite_button = tk.Button(favorites_frame, text="Add to Favorites", command=add_to_favorites, bg="#3498DB", fg="#ECF0F1")
add_favorite_button.pack()

favorite_listbox = tk.Listbox(favorites_frame, selectmode=tk.SINGLE, width=60, height=5, bg="#34495E", fg="#ECF0F1", selectbackground="#3498DB")
for fav_song in favorites:
    favorite_listbox.insert(tk.END, os.path.basename(fav_song))
favorite_listbox.pack()

# Create a frame for volume controls
volume_frame = tk.Frame(root, bg="#2C3E50")
volume_frame.pack(pady=10)

volume_label = tk.Label(volume_frame, text="Volume:", padx=10, pady=5, bg="#2C3E50", fg="#ECF0F1")
volume_label.grid(row=0, column=0)

volume_slider = tk.Scale(volume_frame, from_=0, to=100, orient=tk.HORIZONTAL, command=set_volume, bg="#2C3E50", fg="#ECF0F1")
volume_slider.set(50)  # Set default volume
volume_slider.grid(row=0, column=1, padx=10)

# Run the application
root.mainloop()
```
